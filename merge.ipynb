```python
import pandas as pd
import os
```

```python
df=pd.read_csv("CLHLS_14_withBiomarkers.csv", encoding='latin1')
df
```

```python
unrefined=pd.read_excel("labeledData_unrefined.xlsx")
unrefined
```

```python
# 计算每一列的有效值数量
valid_size = df.notna().sum()

# 将有效值数量填入unrefined的ValidSize列
unrefined['ValidSize'] = unrefined['Variable'].apply(lambda x: valid_size[x] if x in valid_size else None)

unrefined
```

```python
unrefined.to_excel("labeledData_unrefined.xlsx", index=False)
```
