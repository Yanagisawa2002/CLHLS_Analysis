```python
import pandas as pd
import os
```

```python
data=pd.read_csv('CLHLS_14_withBiomarkers.csv', encoding='latin1')
filtered=pd.read_excel('refined.xlsx')
```

```python
# 获取filtered中的Variable列
variables_to_keep = filtered['Variable'].tolist()

# 只保留data中出现在filtered的Variable这一列中的那些列
data = data[variables_to_keep]

data
```

```python

# 定义需要检查缺失值的列
columns_to_check = ['ALB', 'GLU', 'BUN', 'CREA', 'CHO', 'TG', 'GSP', 'CRPHS', 'UA', 'HDLC', 'SOD', 'MDA', 'VD3', 'VITB12', 'UALB', 'UCR', 'UALBBYUCR', 'WBC', 'LYMPH', 'LYMPH_A', 'RBC', 'HGB', 'HCT', 'MCV', 'MCH', 'MCHC', 'PLT', 'MPV', 'PDW', 'PCT']

# 删除在指定列中有缺失值的行
# data[columns_to_check] = pd.to_numeric(data[columns_to_check], errors='coerce')
data = data.dropna(subset=columns_to_check)

# 显示处理后的数据
for i in columns_to_check:
    data[i] = pd.to_numeric(data[i], errors='coerce')
    data[i].fillna(data[i].mean(), inplace=True)
    
data
```

```python
# 将缺失值都填写上2
data = data.fillna(2)
# 将G15开头的列中的所有非2数值都改为-1
g15_columns = [col for col in data.columns if col.startswith('G15')]
for col in g15_columns:
    data[col] = data[col].apply(lambda x: -1 if x != 2 else x)
for col in g15_columns:
    data[col] = data[col].apply(lambda x: 1 if x != -1  else x)
```

```python
# 将ID修改为从1开始计数
data['ID'] = range(0, len(data))
data
```

```python
data.to_csv('slimed_data.csv',index=False)
```
