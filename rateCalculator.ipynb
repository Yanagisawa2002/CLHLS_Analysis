```python
import pandas as pd
import numpy as np
```

```python
'''
# 读取饮食习惯kdm_data.xlsx文件
diet_data = pd.read_excel('饮食习惯kdm_data.xlsx')

# 读取思维清晰kdm_data.xlsx文件
cognition_data = pd.read_excel('思维清晰kdm_data.xlsx')

# 读取娱乐锻炼生活方式kdm_data.xlsx文件
lifestyle_data = pd.read_excel('娱乐锻炼生活方式kdm_data.xlsx')

# 读取体态体能kdm_data.xlsx文件
fitness_data = pd.read_excel('体态体能kdm_data.xlsx')

# 生成一个具有五列的数据帧，这些列分别是diet_data['age']以及余下四个数据帧中各自的'kdm'列
result_df = pd.DataFrame({
    'age': diet_data['age'],
    'diet_kdm': diet_data['kdm'],
    'cognition_kdm': cognition_data['kdm'],
    'lifestyle_kdm': lifestyle_data['kdm'],
    'fitness_kdm': fitness_data['kdm']
})

# 为result_df增加四列，分别计算diet_rate, cognition_rate, lifestyle_rate, fitness_rate
result_df['diet_rate'] = (result_df['diet_kdm'] - result_df['age']) / result_df['age']
result_df['cognition_rate'] = (result_df['cognition_kdm'] - result_df['age']) / result_df['age']
result_df['lifestyle_rate'] = (result_df['lifestyle_kdm'] - result_df['age']) / result_df['age']
result_df['fitness_rate'] = (result_df['fitness_kdm'] - result_df['age']) / result_df['age']

result_df.to_excel('agingRate.xlsx', index=False)
'''
```

```python
# 绘制各个rate的分布图
plt.figure(figsize=(12, 8))


# 绘制认知率分布图
plt.subplot(2, 2, 1)
sns.histplot(agingRate['cognition_rate'], kde=True, color='green')
plt.xlim(-0.5, 1.2)

# 绘制健身率分布图
plt.subplot(2, 2, 2)
sns.histplot(agingRate['fitness_rate'], kde=True, color='purple')
plt.xlim(-0.5, 1.2)

# 绘制饮食率分布图
plt.subplot(2, 2, 3)
sns.histplot(agingRate['diet_rate'], kde=True, color='blue')
plt.xlim(-0.5, 1.2)

# 绘制生活方式率分布图
plt.subplot(2, 2, 4)
sns.histplot(agingRate['lifestyle_rate'], kde=True, color='red')
plt.xlim(-0.5, 1.2)


plt.tight_layout()
plt.show()
```
